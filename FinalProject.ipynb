```python
#Libraries  to read a pdf file
import PyPDF2 as pdf
```

```python
#Open the file
file=open('sample.pdf', 'rb')
print(file)
```

```python
#Create a  object of pdf_reader to read the file
pdf_reader=pdf.PdfFileReader(file)
```

```python
#To check the file is encrypted or not
pdf_reader.getIsEncrypted()
```

```python
#To get number of pages
pdf_reader.getNumPages()
```

```python
#Read all the pages of pdf
pdf_text=""
for i in range(pdf_reader.getNumPages()):
    temp=pdf_reader.getPage(i)
    pdf_text=pdf_text+temp.extractText()
    print(pdf_text)
```

```python
#Correct all the spelling mistake
from textblob import TextBlob
text=TextBlob(pdf_text)
print(text.correct())
```

```python
#Correct all grammertical mistake
from gingerit.gingerit import GingerIt
parser=GingerIt()
result_dict=parser.parse(text)
print(result_dict["result"])
```
